```python
import matplotlib.pyplot as plt
import numpy as np
```

## Types of graphs

1. plot- line graph , between two parameters
2. bar-  bar graph of single parameter (optional categorical data)
3. hist- histogram of a bunch of data i.e. frequency of single parameter
4. scatter- dot graph between two Param
5. boxplot/ box- statstical overview in form of box for a single Param mean, outliers, qunatiles (34%,25%), IQR (optional - categorical data)
6. stem - more like bar graph, lines instead of bars
7. violinplot - violin shaped statistical overview graph spread of data (more like hist) single parameter ( optional categorical data)

```python
x=np.linspace(0,5,11)
y=np.arange(1,11,3)
```

```python
x
```

```python
y
```

```python
y=x**3
plt.plot(x,y)

```

```python
y=x**3
plt.bar(x,y)
```

```python
y=x**3
plt.hist(x,y)
```

```python
y=x**3
plt.scatter(x,y)
```

```python
y=x**3
plt.scatter(x,y)
plt.title('bar graph')
plt.grid()
plt.xlabel('x-->')
plt.ylabel('y-->')
plt.xticks(x)
plt.yticks(range(0,200,10))
plt.show()
```

```python

plt.boxplot(x)
plt.violinplot(x)
```

```python
y2=x**4
```

```python
plt.plot(x,y,x,y2)
plt.legend(['y=x**3','y2=x**4'])
plt.show()
```

```python
fig=plt.figure(figsize=(10,6))
plt.plot(x,y, lw=3, marker='o', markerfacecolor='purple')
plt.plot(x,y2, marker='v', markersize=9,markerfacecolor='purple')
plt.legend(['y=x**3','y2=x**4'])
plt.show()
```
